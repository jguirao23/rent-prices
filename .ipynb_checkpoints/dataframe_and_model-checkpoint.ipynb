```python
import requests
import pandas as pd
import json
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import warnings
warnings.filterwarnings("ignore")
```

Importing the houses.csv

```python
houses = pd.read_csv('houses.csv')
```

```python
houses.columns
```

```python
houses.drop(['Unnamed: 0', 'thumbnail', 'numPhotos', 'showAddress',
       'url', 'distance', 'description', 'hasVideo', 'hasPlan', 'suggestedTexts', 'has3DTour', 'has360',
       'hasStaging', 'labels', 'superTopHighlight', 'topNewDevelopment',
       'externalReference', 'highlight', 'newDevelopmentFinished'], axis=1, inplace=True)
```

```python
houses.head()
```

```python
houses.dropna(subset=['floor', 'exterior'], how='any', inplace=True)
```

```python
houses.head(3)
```

We need to separate the dict values from column parkingSpace into different columns.

```python
houses['isParkingSpaceIncludedInPrice'] = False
houses['hasParkingSpace'] = False
houses['parkingSpacePrice'] = 0.00

for index, rows in houses.iterrows():
    try:
        houses.loc[index, ['isParkingSpaceIncludedInPrice']] =  eval(rows['parkingSpace']).get('isParkingSpaceIncludedInPrice')
        houses.loc[index, ['hasParkingSpace']] =  eval(rows['parkingSpace']).get('hasParkingSpace')
        houses.loc[index, ['parkingSpacePrice']] =  eval(rows['parkingSpace']).get('parkingSpacePrice')
    except:
        pass
```

```python
houses.head()
```

```python
houses.parkingSpacePrice.fillna(0.0, inplace=True)
```

```python
houses['typology'] = False

for index, rows in houses.iterrows():
    try:
        houses.loc[index, ['typology']] =  eval(rows['detailedType']).get('typology')
    except:
        pass
```

```python
houses.typology.unique()
```

```python
houses['totalPrice'] = houses['price'] + houses['parkingSpacePrice']
```

```python
ibiza = houses[houses['neighborhood'] == 'Ibiza']
```

```python
ibiza.drop(['parkingSpace', 'detailedType'], axis=1, inplace=True)
```

```python
ibiza.head()
```

```python
# We want to distinguish if the apartment floor is 'bj' (street level) or not
ibiza['isFloorZero'] = ibiza.floor.apply(lambda x: True if x == 'bj' else False)
```

```python
ibiza.head()
```

### Multiple linear regression model

```python
ibiza.columns
```

```python
ibiza.hasLift = ibiza.hasLift.astype(bool)
ibiza.exterior = ibiza.exterior.astype(bool)
```

```python
cols = ['totalPrice', 'isFloorZero', 'size', 'exterior', 'rooms', 'bathrooms',
       'hasLift', 'hasParkingSpace']
ibiza_model = ibiza[cols]
```

```python
ibiza_model.info()
```

```python
sns.heatmap(ibiza_model.corr(numeric_only=True), annot=True)
```

```python
scaler = StandardScaler()
ibiza_model[['size', 'rooms', 'bathrooms']] = scaler.fit_transform(ibiza_model[['size', 'rooms', 'bathrooms']])
```

```python
ibiza_model.head()
```

```python
sns.heatmap(ibiza_model.corr(numeric_only=True), annot=True)
```

```python
# Multiple linear regression model
X = ibiza_model.drop('totalPrice', axis=1)
y = ibiza_model['totalPrice']

model = linear_model.LinearRegression()
result = model.fit(X, y)
```

```python
result.score(X, y)
```
